# Encoding

```python
num_df = df.select_dtypes(include=['int64','float64']).copy()
num_df
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df
```

```python
obj_df["OHC_Code"] = np.where(obj_df["engine-type"].str.contains("ohc"), 1, 0)
obj_df.head()
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
obj_df["make_code"] = ord_enc.fit_transform(obj_df[["make"]])
obj_df[["make", "make_code"]].head(11)
```

```python
obj_df["FUEL_Code"] = np.where(obj_df["fuel-type"].str.contains("gas"), 1, 0)
obj_df.head()
```

```python
obj_df["engine_loc_Code"] = np.where(obj_df["engine-location"].str.contains("front"), 1, 0)
obj_df.head()
```

```python
obj_df["drive-wheels"].replace('4wd','fwd',inplace=True)
obj_df["drive_wheels_code"] = np.where(obj_df["drive-wheels"].str.contains("rwd"), 1, 0)
obj_df.head()
```

```python
obj_df["body-style"] = obj_df["body-style"].astype('category')
print(obj_df.dtypes)
print(obj_df)
```

```python
obj_df["body_style_cat"] = obj_df["body-style"].cat.codes
obj_df.head()
```
